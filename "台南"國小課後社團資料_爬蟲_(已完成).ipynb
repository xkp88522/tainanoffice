## 針對名單中的所有學校做爬蟲  
目前台南地區共205校，總約花費15-20分鐘

注意: 爬蟲過程中可能有抓取失敗，請再自行另外確認!   (目前只有仁德國小還抓不到)

```python
# 第一種抓取 容器: div , 標籤: h4 中的 a
import requests
from bs4 import BeautifulSoup
import pandas as pd

def crawl_website(url, school_name):
    df = pd.DataFrame(columns=['國小', '標題', '來源網址'])

    full_url = url

    response = requests.get(full_url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # 找到所有文章容器
        articles = soup.find_all('div', class_='news_page_container')

        for article in articles:
            title = article.find('h4').find('a').text.strip()

            # 過濾只保留包含"社團"、"課後"、"照顧"和"才藝"的標題  ## 可更改 ##
            if '社團' in title or '課後' in title or '照顧' in title or '才藝' in title:
                df = df.append({'國小': school_name, '標題': title, '來源網址': full_url}, ignore_index=True)

    else:
        print(f"Error: {response.status_code} for {full_url}")

    return df


original_data = df1   # pd.DataFrame(data)

# 逐一爬蟲每一筆資料
result_dfs = []
for index, row in original_data.iterrows():
    school_name = row['學校名稱']
    url = row['網址']

    # 呼叫爬取函數
    result_df = crawl_website(url, school_name)

    # 如果結果不為空，輸出 DataFrame
    if not result_df.empty:
        result_dfs.append(result_df)

# 合併所有結果
final_result_df1 = pd.concat(result_dfs, ignore_index=True)

# 顯示最終結果
print(final_result_df1)
```
